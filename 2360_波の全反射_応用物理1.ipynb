$\require{cancel}$
### Snell の法則

臨界角 $\theta$ 以下では
\begin{align*}
n_1 \cdot \sin \theta_1& = n_2 \cdot \sin \theta_2&
\end{align*}

臨界角 $\theta_c$ では, 屈折波が境界面を行き来するためエネルギーの変化が抑えられるように, 全反射をする
\begin{align*}
n_1 \cdot \sin \theta_c& = n_2 \cdot \cancelto{1}{\sin \frac{\pi}{2}}&\\
\sin \theta_c &= \dfrac{n_2}{n_1}
\end{align*}

正弦関数 $\sin$ の上限を考えれば
\begin{align*}
 \dfrac{n_2}{n_1} &= \sin \theta_c <1&\\
  \dfrac{n_2}{n_1} &<1&\\
\end{align*}

波の速さ$v_1, v_2$で屈折の法則をせいりすると
\begin{align*}
  \dfrac{n_2}{n_1} = \dfrac{v_1}{v_2} &<1&\\
  v_1 &< v_2&
\end{align*}
つまり全反射は, 
>> 元の媒質から異なる物質に射線が進むとき, 波の速さが大きくなるときに起こる

ということができます.
