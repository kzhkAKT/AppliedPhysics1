1. **強め合う条件** :: 時間変化で **激しく振動する** 条件 ~ 定常波の **腹**
2. *弱め合う合う条件* :: 時間変化で *全く振動しない* 条件 ~ 定常波の *節*

- 波源 = source of wave -> S
- 観察点 = a point -> P or Q

簡単のために、これから考える干渉については, **二つの波源の位相は等しい** ものとします。

このような条件を
> 波源は同位相である

と言います.

```
#@title 各波源と観測点との差 $\Delta L$
%%html
<iframe src="https://www.desmos.com/geometry/szx6vea0jy?embed" width="500" height="500" style="border: 1px solid #ccc" frameborder=0></iframe>
```

$\require{cancel}$
#### 強め合う条件

>> ある点 P について, 二つの同位相の波源 ${\rm S_1, S_2}$からの距離の差 $\Delta L$ が **波長 $\lambda$ の整数倍**  $m$ であれば, 強め合う(腹になる)

\begin{align*}
\Delta L &= \left|\overline{\rm S_1P} - \overline{\rm S_2P}\right|
 = m \lambda &\\
 &= \bcancel{2} m \dfrac{\lambda}{\bcancel{2}} = 2 m \dfrac{\lambda}{2}&
\end{align*}

この関係式を
> 波源が同位相で $\Delta L$ が 半波長 $\dfrac{\lambda}{2}$ の **偶数倍** のとき, 注目する点は腹になる

と読むこともできます.



#### 弱め合う条件

>> ある点 Q について, 二つの同位相の波源 ${\rm S_1, S_2}$からの距離の差 $\Delta L$ が波長 $\lambda$ の整数倍 $m$ から **半波長 $\dfrac{\lambda}{2}$ ずれていれば**,  弱め合う(節になる)

\begin{align*}
\Delta L &= \left|\overline{\rm S_1Q} - \overline{\rm S_2Q}\right|
 = m \lambda + \dfrac{\lambda}{2}= \left( m + \dfrac{1}{2}\right)\lambda&\\
 &= \left(\bcancel{2} m + 1 \right)\dfrac{\lambda}{\bcancel{2}}
 = \left(2 m +1 \right) \dfrac{\lambda}{2}&
\end{align*}

この関係式を
> 波源が同位相で $\Delta L$ が 半波長 $\dfrac{\lambda}{2}$ の **奇数倍* のとき, 注目する点は節になる

と読むこともできます.

### 波の干渉と双曲線

強め合っている点も弱めあっている点も次の関係を満たしています:
\begin{align*}
\Delta L &= \left|\overline{\rm S_1Q} - \overline{\rm S_2Q}\right|
\end{align*}
ここでは, 分かりするために近接する弱めあっている点(節)に注目しましょう.
