## 正弦波の導出: 原点の振動の遅れ観察 

```
#@title 原点の振動の遅れ観察 原点の振動の遅れ観察 : (Slow Mothon + Pulse + No End + Damping:None)
: (Slow Mothon + Pulse + No End + Damping:None)
%%html
<iframe src="https://phet.colorado.edu/sims/html/wave-on-a-string/latest/wave-on-a-string_en.html" width="720" height="540" scrolling="no" allowfullscreen></iframe>
```

原点の振動 $y_0$ を 振幅$A$, 周期$T$, 振動数$f$ の単振動とする
\begin{align*}
y_0 &= A \sin \omega t = A \sin \dfrac{2\pi}{T} t = A \sin 2\pi f t &\\
&= y_0(t)&
\end{align*}

この原点 O の振動が, 
- $x$軸 の正の向きに (一定の)速さ $v$ で伝わるとき
- 原点 O から $x$ だけ離れた点P における 正弦波の変位 $y$ を求める.

ポイントは二つ

1. 原点の振動が点P伝わるまでに伝わる時間: $\quad t' = \dfrac{x}{v}$
2. 点Pでの変位は, 原点Oの変位から遅れる: $\quad t-t'$

これらを原点の単振動へ組み込むと,(波の基本式:$v=\frac{\lambda}{T}$)
\begin{align*}
y &= y_0(t-t') = A \sin \dfrac{2\pi}{T} \left( t- \dfrac{x}{v}\right)&\\
&= A \sin 2\pi\left( \dfrac{t}{T}- \dfrac{x}{vT}\right)&\\
&= A \sin 2\pi\left( f t - \dfrac{x}{\lambda}\right)&\\
\end{align*}

より一般化し, 原点での振動が初期位相 $\alpha$ を持ち $x$ 軸上を正の方向に伝播する正弦波の変位は次のようになります.
\begin{align*}
y &= A \sin \left\{ 2\pi\left(f t - \dfrac{x}{\lambda}\right)+ \alpha\right\}&\\
\end{align*}

- 位相: 括弧 $\{\}$ の中. 媒質の振動の情報をあたえます.



> Q. 上の解説を参考に, $x$軸上を **負の方向に** 伝播する正弦波の変位を示してください.

## 正弦波の導出: 波形の拡大と平行移動

```
#@title y=sin(x);  y=sin(x+2π)=sin(x)
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 1000)

fig = plt.figure()
plt.xlim(0, 10)
plt.ylim(-5,+5)
plt.axhline(y=0, color='k',linestyle='dashed', linewidth=0.5)
plt.plot(x, np.sin(x), '-')
```

```
#@title y= sin(2πx); y=sin(2π(x+1))= sin(2πx+2π) = sin(2πx)
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 1000)

A=1

fig = plt.figure()
plt.xlim(0, 10)
plt.ylim(-5,+5)
plt.axhline(y=0, color='k',linestyle='dashed', linewidth=0.5)
plt.plot(x, A*np.sin(2*np.pi*x), '-')
```

```
#@title y= sin(2πx/λ), λ=3; y= sin(2π(x+λ)/λ) = sin(2πx/λ+2π) = sin(2πx)
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 1000)

A=1
lmd=3

fig = plt.figure()
plt.xlim(0, 10)
plt.ylim(-5,+5)
plt.axhline(y=0, color='k',linestyle='dashed', linewidth=0.5)
plt.plot(x, A*np.sin(2*np.pi*x/lmd), '-')
```

```
#@title y=A sin(2πx/λ), A=4 m, λ= 3 m; y= A sin(2π(x+λ)/λ) = A sin(2πx/λ+2π) = A sin(2π/λ x)
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 1000)

A=4
lmd=3

fig = plt.figure()
plt.xlim(0, 10)
plt.ylim(-5,+5)
plt.axhline(y=0, color='k',linestyle='dashed', linewidth=0.5)
plt.plot(x, A*np.sin(2*np.pi*x/lmd), '-')
```
