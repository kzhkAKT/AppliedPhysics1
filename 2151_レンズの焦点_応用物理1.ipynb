### レンズの種類

#### 用語確認

教科書 p96 図2.12 を見ながら

- 光軸: 半径 $R_1, R_1$ の中心 を結ぶ線
- 近軸光線: 光軸の近傍を通る光線
- 曲率半径: レンズを作る半径 $R_1, R_1$ 
- 焦点: レンズを通った光が集まる点
- 焦点距離: レンズの中心から焦点までの距離 $f$
- 薄肉レンズ: レンズの厚みが $R_1, R_2$ より十分小さい

##### レンズの前後

- レンズの前方: レンズから見て光が入射する方向
- レンズの後方: レンズから見て光が出て行く方向




### レンズの公式

プリズムの問題(DG.385,389) を利用することで導出できます。  

導出を簡単にするために凸レンズのみを考え, レンズは薄肉とします.
+ 凸レンズのみを考え
+ レンズは薄肉
+ 近軸近似
を適用します. これらを踏まえると近似公式
> $\qquad \sin \theta \approx \theta \approx \tan \theta$

が利用できます.

#### プリズムの製作

\end{align*}
が導出できます.

#### プリズムを利用した屈折

レンズの中心(M)からレンズの前方の光軸 $a$ から出発した光線が,
レンズの後方の光軸 $b$ に達したとします. この光線が光軸となす角を下図のようにとれば
\begin{align*}
\delta \approx \dfrac{h}{a} + \dfrac{h}{b}
\end{align*}


#### レンズの公式の導出
これらを自学自習課題の問題を利用して整理すると
$\require{cancel}$
\begin{align*}
\delta &= \left(n-1\right)\theta&\\
\dfrac{\cancel{h}}{a} + \dfrac{\cancel{h}}{b}
&\approx
\left(n-1\right)
\left( \dfrac{\cancel{h}}{R_1} + \dfrac{\cancel{h}}{R_2} \right)&\\
\dfrac{1}{a} + \dfrac{1}{b}
&\approx
\left(n-1\right)
\left( \dfrac{1}{R_1} + \dfrac{1}{R_2} \right)&\\
\end{align*}
ここで, 凸レンズであれば光線はレンズ後方で光軸へ戻るので, これを $\delta >0 $とします.   
さらに $a \to \infty$ のとき $b \to f$ と一点に集まるので
\begin{align*}
\cancelto{0}{\dfrac{1}{a}} + \dfrac{1}{\cancelto{f}{b}}
&\approx
\left(n-1\right)
\left( \dfrac{1}{R_1} + \dfrac{1}{R_2} \right)&\\
\dfrac{1}{f} & = \left(n-1\right)
\left( \dfrac{1}{R_1} + \dfrac{1}{R_2} \right)&\\
\end{align*}
となります.
