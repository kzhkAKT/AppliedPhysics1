#### 凸レンズの実像の倍率と物体と実像の関係式

**実像** とは光源が実際に集まって作る像のことです.

凸レンズの **実像** は, 図で確認できるように倒立しているので,
**倒立像** と言います.

##### 倍率
+ 物体と実像は(光源と像なので)平行(緑矢印)
+ (対称な)レンズの真ん中を通る光線は影響を受けず直進

これから, 倍率 $m$ を図形的な条件を求めると,

> $\qquad m = \dfrac{h_{\rm 像}}{h_{\rm 物体}} = \dfrac{S_2}{S_1}$

##### 物体と実像の関係式
+ 物体から出た光軸に平行な光
+ レンズの中心に物体と同じ高さが取れる

これから, 倍率 $m$ を図形的な条件を求めると,

> $\qquad m = \dfrac{h_{\rm 像}}{h_{\rm 物体}} = \dfrac{S_2 - f}{f}$

これらをまとめると
\begin{align*}
\dfrac{S_2}{S_1} &= m = \dfrac{S_2 - f}{f}&\\
S_2 f &= S_1 S_2 - f S_1&\\
S_2 f + f S_1 &= S_1 S_2 &
\end{align*}
ここで両辺を $S_1 S_2 f$ で割ると
\begin{align*}
\frac{S_2 f}{S_1 S_2 f} + \frac{f S_1}{S_1 S_2 f}
&= \frac{S_1 S_2}{S_1 S_2 f} &\\
\end{align*}

>> $\qquad \dfrac{1}{S_1} + \dfrac{1}{S_2} = \dfrac{1}{f}$
