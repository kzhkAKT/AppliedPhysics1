## レンズ

+ (対称な)レンズの真ん中を通る光線は影響を受けず直進
+ 光軸に平行な光の行く先
+ 焦点を通過/焦点へ延長される光線  

に注意し **図形の相似** を考えれば以下の式は導出できます.

### 凸レンズによる虚像

像が延長された光線から *あたかも* 届いたように捉えることができるため, このような像を **虚像** という.

レンズ後方の像が, レンズ前方にやってきたことから, $S_2 \to -S_2$ と読み替えることでそのまま先ほどの式が利用できる.

##### 倍率

> $\qquad m = \dfrac{h_{\rm 像}}{h_{\rm 物体}} = \dfrac{S_2}{S_1}$

##### 物体虚像の関係式

レンズ後方の焦点を頂点とした相似を考慮すると

> $\qquad m = \dfrac{h_{\rm 像}}{h_{\rm 物体}} = \dfrac{S_2+f}{S_1}$

これを整理すると

>> $\qquad \dfrac{1}{S_1} + \dfrac{1}{-S_2} = \dfrac{1}{f}$
