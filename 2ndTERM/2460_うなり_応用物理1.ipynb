### 440Hz

```
#@title A=440Hz: y-t グラフ
% matplotlib inline
import matplotlib.pyplot as plt
#https://colab.research.google.com/drive/1--xY78_ZTFwpI7F2ZfaeyFKiAOG2nkwd
# Needed imports
import numpy as np
from IPython.display import Audio
from scipy.io import wavfile

# Generate a sound
import numpy as np
framerate = 44100
T = 2

f_A = 440
# f_2A = 2*f_A

t0 = np.linspace(0, T, framerate*T)
data0 = np.sin(2*np.pi*f_A*t0)

plt.plot(t0, data0, '-c', label='440 Hz in [0,0.10] s')
plt.legend(loc='upper right')
plt.xlim(0, 1.0/10)
plt.xlabel("$t$ [s]")
plt.ylabel("$A$ ")
plt.show()
```

```
#@title 時間拡大 440Hz
plt.plot(t0, data0, '-c', label='440 Hz in [0,0.001] s')
plt.legend(loc='upper right')
plt.xlim(0, 1.0/100)
plt.xlabel("$t$ [s]")
plt.ylabel("$A$ ")
plt.show()
```

```
#@title 440 Hz 再生
Audio(data0,rate=framerate)
```

### (440-10) Hz

```
#@title A=(440-10)Hz: y-t グラフ
% matplotlib inline
import matplotlib.pyplot as plt
#https://colab.research.google.com/drive/1--xY78_ZTFwpI7F2ZfaeyFKiAOG2nkwd
# Needed imports
import numpy as np
from IPython.display import Audio
from scipy.io import wavfile

# Generate a sound
import numpy as np
framerate = 44100
T = 2

f_A = 440

df=10
f_Am = f_A - df

t1 = np.linspace(0, T, framerate*T)
data1 = np.sin(2*np.pi* f_Am *t1)

plt.plot(t1, data1, '-c', label='(440-30) Hz in [0,0.10] s')
plt.legend(loc='upper right')
plt.xlim(0, 1.0/10)
plt.xlabel("$t$ [s]")
plt.ylabel("$A$ ")
plt.show()
```

```
Audio(data1,rate=(framerate))
```

```
#@title sound:{440 Hz} と sound:{(440-10) Hz} の干渉(~重ね合わせ): うなり
t2 = np.linspace(0, T, framerate*T)
data2 = data0 + data1

plt.plot(t2, data2, '-c', label='sound:{440 Hz} +sound:{(440-10) Hz} in [0,0.001] s')
plt.legend(loc='upper right')
plt.xlim(0, 1.0/100)
plt.xlabel("$t$ [s]")
plt.ylabel("$A$ ")
plt.show()
```

```
#@title 上のうなりを時間幅を変えて観察
plt.plot(t2, data2, '-c', label='440+(440-20) Hz in [0,0.01] s')
plt.legend(loc='upper right')
plt.xlim(0, 1.0/10)
plt.xlabel("$t$ [s]")
plt.ylabel("$A$ ")
plt.show()
```

### 2. 簡単な音波の合成



#### 2.1
\begin{align*}
y_1 &= A \sin \left( 2\pi f_1 t \right)&\\
y_2 &= A \sin \left( 2\pi f_2 t \right)&\\
\end{align*}

合成音 $y_1 + y_2$ を和の形から、積の形に変換せよ.
ただし

\begin{align*}
f_1 &= \dfrac{f_1 + f_2}{2} + \dfrac{f_1 - f_2}{2} &\\
f_2 &= \dfrac{f_1 + f_2}{2} - \dfrac{f_1 - f_2}{2} 
\end{align*}

を利用して良い

$\require{cancel}$

解答: 2.1
\begin{align*}
y_1 + y_2 &= A \sin 2\pi f_1 t  + A \sin 2\pi f_2 t&\\
=& A \sin 2\pi t\left( \dfrac{f_1 + f_2}{2} + \dfrac{f_1 - f_2}{2}\right)
+  A \sin 2\pi t\left( \dfrac{f_1 + f_2}{2} - \dfrac{f_1 - f_2}{2}\right) &\\
=& 
A 
\sin 2\pi t\left( \dfrac{f_1 + f_2}{2}\right) 
\cos 2\pi t\left(\dfrac{f_1 - f_2}{2}\right)
+
\bcancel{\cos 2\pi t\left( \dfrac{f_1 + f_2}{2}\right) 
\sin 2\pi t\left(\dfrac{f_1 - f_2}{2}\right)}&\\
&+
\sin 2\pi t\left( \dfrac{f_1 + f_2}{2}\right) 
\cos 2\pi t\left(\dfrac{f_1 - f_2}{2}\right)
-
\bcancel{\cos 2\pi t\left( \dfrac{f_1 + f_2}{2}\right) 
\sin 2\pi t\left(\dfrac{f_1 - f_2}{2}\right)}&\\
=&
2A \sin 2\pi t\left( \dfrac{f_1 + f_2}{2}\right) 
\cos 2\pi t\left(\dfrac{f_1 - f_2}{2}\right)&
\end{align*}

#### 2.3

\cos \pi t\left(f_1 - f_2\right)&
\end{align*}

## 復習: うなり

- 振動数の近い二つのおんさを同時に鳴らすと, 音の強弱が発生し, うなりが発生する
- 音は縦波/疎密波である事を思い出すと,
  - おんさ一つを鳴らしたときは, 空気の分子の密度変化が連続的に起こり, 唸りは聞こえない
  - 振動数の近い二つのおんさを同時に鳴らしたときは, 空気の連続的な密度変化が起こりながらも,   
  重ね合わせの原理により合成波の振幅が大きい場所と全く振動しない点が出来上がる
- 振動数の近い二つのおんさの振動数をそれぞれ $f, f'$ で うなりが 1秒間に $n$ 回聞こえたときは, 次の関係式を満たす
  \begin{align*}
  n = \left| f - f' \right|
  \end{align*}
