```
#@title A=440Hz A-t グラフ
% matplotlib inline
import matplotlib.pyplot as plt
#https://colab.research.google.com/drive/1--xY78_ZTFwpI7F2ZfaeyFKiAOG2nkwd
# Needed imports
import numpy as np
from IPython.display import Audio
from scipy.io import wavfile

# Generate a sound
import numpy as np
framerate = 44100
T = 2

f_A = 440
# f_2A = 2*f_A

t0 = np.linspace(0, T, framerate*T)
data0 = np.sin(2*np.pi*f_A*t0)

plt.plot(t0, data0, '-c', label='440 Hz in [0,0.10] s')
plt.legend(loc='upper right')
plt.xlim(0, 1.0/10)
plt.xlabel("$t$ [s]")
plt.ylabel("$A$ ")
plt.show()
```

```
#@title 時間拡大
plt.plot(t0, data0, '-c', label='440 Hz in [0,0.010] s')
plt.legend(loc='upper right')
plt.xlim(0, 1.0/50)
plt.xlabel("$t$ [s]")
plt.ylabel("$A$ ")
plt.show()
```
