#### ホイヘンスの原理

```
#@title 入射角 と 屈折角 + 法線
%%svg
https://upload.wikimedia.org/wikipedia/commons/3/3f/Snells_law2.svg
```

#### スネルの法則

$\require{cancel}$

真空中の屈折率 $n_1$で, 媒質の絶対屈折率を $n_2$ とすると
\begin{align}
{n_1} \sin \cancelto{i}{\theta_1}
&= {n_2} \sin \cancelto{r}{\theta_2}\\
\dfrac{\sin i}{\sin r}&=\dfrac{n_2}{n_1} = n,
\end{align}
これまでの議論であれば, $n$ は真空に対する相対屈折率ですが, 
光速 $c$ を基準にするときに改めて **絶対屈折率** とします.


これは次の様に考えると良いでしょ
- 光速 $c$ : $n_1\to1$
- 媒質の屈折率 : $n_2\to n$
\begin{align}
\dfrac{\sin i}{\sin r}
&=\dfrac{\cancelto{n}{n_2}}
{\cancelto{1}{n_1}} = n,
\end{align}

#### ホイヘンスの原理の適用

- 真空: 速度 $=$ 光速: $c$, 入射角 $i$, 波長 $\lambda$
- 媒質中: 速度 $\lneqq$ 光速: $v$, 屈折角 $r$, 波長$\lambda'$
  - 媒質中では, 他の原子や分子が多く存在し邪魔者が多く早く進めないと考えてください
\begin{align}
n&=\dfrac{\sin i}{\sin r}
=\dfrac{L\sin i}{L\sin r}\\
&= \dfrac{\lambda}{\lambda'}
= \dfrac{\lambda f}{\lambda' f} \\
&= \dfrac{c}{v}
\end{align}
これを改めてまとめると
\begin{align}
\dfrac{\sin i}{\sin r} = n = \dfrac{c}{v} > 1
\end{align}

#### 光学的な疎密

絶対屈折率 $n, n'$ の媒質が
\begin{align}
n' > n
\end{align}
を満たすとき
- 大きい $n'$ を光学的に密な媒質
- 小さい $n$ を光学的に疎な媒質

と言います. 
この条件のもので
> $n'$ の媒質から $n$ の媒質に光が入射するとき,
臨界角 $i_c$ で **全反射** が起こります
